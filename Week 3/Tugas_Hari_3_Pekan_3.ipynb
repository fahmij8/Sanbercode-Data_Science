```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15,8), sharey=True)
ax[0].plot(x, tan, color='b', label='tanX')
ax[0].set_title('Grafik tanx')
ax[1].plot(x, sin, color='r', label='sinX')
ax[1].set_title('Grafik sinx')
ax[2].plot(x, cos, color='g', label='cosX')
ax[2].set_title('Grafik cosx')
fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
ax.plot(x, y, color='#3483BA')
ax.set_title('Grafik CosX')

fig2 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
fig2.plot(x, y2, color='#FF0000')
fig2.plot(x, y, color='#0000FF')
fig2.set_title('Grafik CosX^2')

fig3 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
fig3.plot(x, y3, color='#42A142')
fig3.plot(x, y, color='#0000FF')
fig3.set_title('Grafik CosX^3')

fig4 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
fig4.plot(x, y4, color='#000000')
fig4.plot(x, y, color='#0000FF')
fig4.set_title('Grafik CosX^4')

fig5 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
fig5.plot(x, y5, color='#C6C61F')
fig5.plot(x, y, color='#0000FF')
fig5.set_title('Grafik CosX^5')
plt.show()
```
